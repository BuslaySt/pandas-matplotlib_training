```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('GooglePlayStore_wild.csv')

def set_season(date):
    '''Set new column Season'''
    month = date.split()[0]
    seasons = {'Winter': ['December', 'January', 'February'],
                'Spring': ['March', 'April', 'May'],
                'Summer': ['June', 'July', 'August'],
                'Autumn': ['September', 'October', 'November']}
    for season in seasons:
        if month in seasons[season]:
            return season
 
df['Season'] = df['Last Updated'].apply(set_season)
```

```python
s = df.pivot_table(index = 'Season', columns = ['Type', 'Content Rating'], values = 'Rating', aggfunc = 'median')
s.drop(columns=['Unrated','Adults only 18+'], level=1, inplace = True)
```

```python
s.plot(kind = 'barh', subplots = True, layout = (2,4), grid = True, figsize = (18, 15), sharey = True, legend = False)
plt.show()
```
